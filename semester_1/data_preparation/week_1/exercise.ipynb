```python
low_df = df.loc[df['cut']=='Ideal']
low_df.head()
```

```python
df['price_per_carat'] = df['price']/df['carat']
df['ppc_high'] = np.where(df['price_per_carat'] > 3500, 1, 0)
df.head()
```

```python
df['rounded_price'] = df['price'].apply(math.ceil)
df['rounded_price_100'] = df['price'].apply(lambda x: math.ceil(x/100)*100)
df.head()
```

```python
def get_100_multiple_ceil(x):
    return math.ceil(x/100)*100

df['rounded_price_100'] = df['price'].apply(lambda x: math.ceil(x/100)*100)
df.head()
```

```python
def is_desired(x):
    return 'yes' if (x['cut'] == 'Ideal' and x['color'] == 'D') else 'no'

df['desired'] = df.apply(is_desired, axis=1)
df.head()
```

```python
df = df.drop(columns=['rounded_price', 'rounded_price_100'])
df.head()
```
